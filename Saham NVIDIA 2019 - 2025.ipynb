Berdasarkan grafik di atas, daily return pada saham `NVIDIA` mengalami volatilitas yang tinggi pada tahun 2020 awal dan pertengahan 2025.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

df_reg = df
df_reg['SMA_20'] = df_reg['Close'].rolling(window=20).mean()
df_reg['Lag_1'] = df_reg['Close'].shift(1)
df_reg['Daily_Return'] = df_reg['Close'].pct_change()

# Drop baris yang kosong
df_reg.dropna(inplace=True)

# Persiapan data untuk Linear Regression
X = df_reg[['Lag_1', 'SMA_20', 'Volume']]
y = df_reg['Close']

# Split data: 80% training, 20% testing
split = int(0.8 * len(df_reg))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Buat model dan training
model = LinearRegression()
model.fit(X_train, y_train)

# Prediksi
y_pred = model.predict(X_test)

# Evaluasi model
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print(f"MAPE: {mape:.2f}%")
print(f"MAE: {mae}")
print(f"RMSE: {rmse}")

# Visualisasi hasil prediksi
plt.figure(figsize=(12, 6))
plt.plot(df_reg['Date'][split:], y_test, label='Actual Price', color='green')
plt.plot(df_reg['Date'][split:], y_pred, label='Predicted Price', color='red')
plt.title('Actual vs Predicted Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid()
plt.show()
```

```python
future_predictions = []
last_known = df_reg.iloc[-1].copy()

for i in range(30):
    lag_1 = last_known['Close']
    
    # Update SMA_20
    sma_20 = df_reg['Close'].iloc[-19:].tolist() + future_predictions  # ambil 19 nilai terakhir + prediksi sebelumnya
    sma_20 = np.mean(sma_20[-20:])
    
    # Ambil volume terakhir (anggap tidak berubah)
    volume = last_known['Volume']
    
    # Buat input prediksi hari ke-i
    X_future = np.array([[lag_1, sma_20, volume]])
    y_future = model.predict(X_future)[0]
    
    # Simpan hasil
    future_predictions.append(y_future)
    
    # Update last_known untuk prediksi berikutnya
    last_known['Close'] = y_future  # hasil jadi lag_1 berikutnya

# Tampilkan hasil prediksi
for i, price in enumerate(future_predictions, start=1):
    print(f"Prediksi Hari ke-{i}: {price:.2f}")
future_dates = pd.date_range(start=df_reg['Date'].iloc[-1] + pd.Timedelta(days=1), periods=30, freq='B')
plt.figure(figsize=(10,5))
plt.plot(df_reg['Date'], df_reg['Close'], label='Historical')
plt.plot(future_dates, future_predictions, label='Forecast 30 Hari', color='red', marker='o')
plt.title('Forecast Linear Regression 30 Hari')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```
