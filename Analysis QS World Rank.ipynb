```python
df = df[df['Overall_Score'] != 0]
numeric_cols = [
    'Academic_Reputation_Score', 'Employer_Reputation_Score', 'Faculty_Student_Score',
    'Citations_per_Faculty_Score', 'International_Faculty_Score',
    'International_Students_Score', 'International_Research_Network_Score',
    'Employment_Outcomes_Score', 'Sustainability_Score', 'Overall_Score'
]

for col in numeric_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True, color='teal')
    plt.title(f'Distribusi {col}', fontsize=14)
    plt.xlabel(col)
    plt.ylabel('Frekuensi')
    plt.show()
```

```python
categorical_cols = ['Region', 'SIZE', 'FOCUS', 'STATUS']

for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, palette='viridis', order=df[col].value_counts().index)
    plt.title(f'Distribusi {col}', fontsize=14)
    plt.xticks(rotation=45)
    plt.ylabel('Jumlah Universitas')
    plt.show()
```

```python
for cat_col in categorical_cols:
    plt.figure(figsize=(12, 8))
    sns.boxplot(data=df, x=cat_col, y='Overall_Score', palette='Set2')
    plt.title(f'Perbandingan Overall Score berdasarkan {cat_col}', fontsize=14)
    plt.xticks(rotation=45)
    plt.show()
```

```python
plt.figure(figsize=(14, 12))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title('Matriks Korelasi', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(
    data=df[df['Overall_Score'] != 0],
    x='International_Research_Network_Score',
    y='Overall_Score',
    hue='Region',
    palette='tab10',
    alpha=0.7
)
plt.title('Hubungan International Research Network dan Overall Score', fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```

```python
top10 = df.sort_values(by='Overall_Score', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(data=top10, y='Institution_Name', x='Overall_Score', palette='coolwarm')
plt.title('Top 10 Universitas Dunia Berdasarkan Overall Score', fontsize=14)
plt.xlabel('Overall Score')
plt.ylabel('Nama Universitas')
plt.show()
```

```python
# Menghitung perubahan peringkat
df['Rank_Change'] = df['RANK_2024'] - df['RANK_2025']

# Mengambil 15 universitas dengan kenaikan peringkat tertinggi
top_improvers = df.sort_values(by='Rank_Change', ascending=False).head(15)

plt.figure(figsize=(12, 8))
sns.barplot(
    x='Rank_Change',
    y='Institution_Name',
    data=top_improvers,
    palette='summer' # Palet warna hijau-kuning
)
plt.title('15 Universitas dengan Kenaikan Peringkat Tertajam (2024-2025)', fontsize=16, fontweight='bold')
plt.xlabel('Jumlah Kenaikan Peringkat', fontsize=12)
plt.ylabel('Nama Institusi', fontsize=12)
plt.show()
```

```python
# Mengambil 10 universitas teratas
top_10_df = df.head(10)

# Memilih kolom skor untuk heatmap
score_columns = [
    'Academic_Reputation_Score', 'Employer_Reputation_Score',
    'Faculty_Student_Score', 'Citations_per_Faculty_Score',
    'International_Students_Score', 'Sustainability_Score', 'Overall_Score'
]

# Membuat heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(
    top_10_df.set_index('Institution_Name')[score_columns],
    annot=True,        # Menampilkan angka di dalam kotak
    cmap='YlGnBu',     # Palet warna dari kuning-hijau-biru
    linewidths=.5,     # Garis antar sel
    fmt='.1f'          # Format angka (1 desimal)
)
plt.title('Heat Maps Kinerja 10 Universitas Terbaik Dunia', fontsize=16, fontweight='bold')
plt.xlabel('Metrik Kinerja', fontsize=12)
plt.ylabel('Nama Institusi', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Memilih semua kolom skor untuk korelasi
correlation_cols = df.select_dtypes(include=np.number).filter(like='Score').columns

# Menghitung matriks korelasi
corr_matrix = df[correlation_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(
    corr_matrix[['Overall_Score']].sort_values(by='Overall_Score', ascending=False),
    annot=True,
    cmap='coolwarm', # Palet merah-biru untuk korelasi positif dan negatif
    linewidths=.5,
    fmt='.2f'
)
plt.title('Korelasi Overall Score dengan Seluruh Score', fontsize=16, fontweight='bold')
plt.show()
```
